3. From the data given below,
  - break the data in to pieces from 0 t0 2 ,3 t0 7
  - concatenate the pieces.
  - concatenate the pieces along axis 1
  - label each as 'a' and 'b' respectively and perform concatenation.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(10, 4))
df
#please provide your answer below this line
```

```python
#break
df3 = df.iloc[0:3]
df4 = df.iloc[3:8]
df3,df4
```

```python
#concat
df5 = pd.concat([df3,df4])
df5
```

```python
#along axis
df6 = pd.concat([df3,df4], axis=1)
df6
```

```python

#label 'a' and 'b',concat
df7 = pd.concat([df3,df4], keys=["a","b"])
df7
```
